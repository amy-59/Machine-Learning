```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples= 100, n_features=1, n_informative=1, n_targets= 1, noise=20, random_state=13)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

### Ridge regression 

```python
from sklearn.linear_model import Ridge
```

```python
ridge1 = Ridge(alpha=10)
ridge1.fit(X,y)
print(ridge1.coef_)
print(ridge1.intercept_)
```

```python
ridge2 = Ridge(alpha=100)
ridge2.fit(X,y)
print(ridge2.coef_)
print(ridge2.intercept_)
```

```python
plt.plot(X,y,'b.')
plt.plot(X,lr.predict(X),color='red', label='alpha = 0')
plt.plot(X,ridge1.predict(X),color='green', label='alpha = 10')
plt.plot(X,ridge2.predict(X),color='orange', label='alpha = 100')
plt.legend()
plt.show()
```
